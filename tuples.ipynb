**Nested Tuples**

```python
t = (1,2,(1,2,3))
print(t)
```

**Type Function**

```python
print(type(t))
```

**Declaring the tuple with one element**

```python
t1 = ('tuples')
print(type(t1))

# Must add , after the element to declar a one element tuple
t1 = ('tuples',)
print(type(t1))
```

**Index use**

```python
print(t[1])
```

**index slicing**

```python
print(t[1:4])
```

**Negaive Index**

```python
print(t[:-2])
```

**Ading two tuples with + operator**

```python
t= (1,2,3) +(4,5,6)
print(t)
```

**Deletion of the tuple**

```python
del t
```

**Tuple membership**

```python
t =(1,2,3,4,5,6,7)
print(4 in t)
print(4 not in t)
```

**Sum, Min and Max function**

```python
print(max(t))
print(sum(t))
print(min(t))
```
